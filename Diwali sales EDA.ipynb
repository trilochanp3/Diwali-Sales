```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Diwali Sales Data.csv",encoding="unicode_escape")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status':'Shaadi'})
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

# Gender

```python
sns.countplot(data=df,x='Gender',hue='Gender')
```

```python
ax= sns.countplot(data=df,x='Gender',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
a = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=a)
```

```python
# From above graphs we can conclude that most of the buyers are females and most amount spent is also by females
```

# Age

```python
df.columns
```

```python
# first using countplot we will find how many people then using barplot we find how much amount
```

```python
sns.countplot(x='Age Group',data=df)  
```

```python
agec = sns.countplot(x='Age Group',hue='Gender',data=df)  
for bars in agec.containers:
    agec.bar_label(bars)
```

```python
total_ages = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group', y='Amount',data = total_ages)
```

```python
# From above graphs we can conclude most of the buyers are from age group 26-35
```

# States

```python

stateo = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
sns.barplot(x='State',y='Orders',data=stateo)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
stateo = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.barplot(x='State',y='Orders',data=stateo)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
statea = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(x='State',y='Amount',data=statea)
```

```python
# From above graphs we can conclude that most of the sales with respect to order and amount are from UP, maharashtra, karnataka, delhi
```

# Marital Status

```python
sns.set(rc={'figure.figsize':(5,5)})
ms = sns.countplot(x='Marital_Status',data=df)
for bars in ms.containers:
    ms.bar_label(bars)
```

```python
ma = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data=ma)
```

```python
# from above graphs we can cocnclude most expenditure is done by married women
```

# Occupation

```python
sns.set(rc={'figure.figsize':(15,5)})
oc = sns.countplot(data=df,x='Occupation')
for bars in oc.containers:
    oc.bar_label(bars)
```

```python
oca = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation',y='Amount',data=oca)
```

```python
# from above graphs we can conclude that most expense with respect to occupation is done by IT sector, Healthcare, Aviation
```

# Product category

```python
sns.set(rc={'figure.figsize':(25,5)})
oc = sns.countplot(data=df,x='Product_Category')
for bars in oc.containers:
    oc.bar_label(bars)
```

```python
pca= df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(data=pca, x='Product_Category',y='Amount')
```

```python
# From above graphs we can conclude mose sold items are food, clothing, electronics, footwear
```

```python
pca= df.groupby(['Product_ID'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(data=pca, x='Product_ID',y='Amount')
```

# Conclusion : 

```python
# Married women between age 26-35  from UP, Maharashtra and Karnataka have purchased more items in terms of count and amount from Food, Clothing,
#Electronics and, Footwear sector
```
